```python
def create_segment_filter(platform, users_list):
    filter_obj_array = []
    for uid_dimension in user_id_variations[platform]:
        uid_filter = Filter( type="in", \
                  dimension=uid_dimension, \
                  values=users_list  \
                 )
        filter_obj_array.append(uid_filter)
    vip_filter = Filter(
                        type="or",
                        fields=filter_obj_array
                        )
    
    return vip_filter
        

def get_valid_userId(row, platform):
    return [ int(v) for v in row[user_id_variations[platform]] if v not in [0, None] and str(v).isdigit() ][0]
    
def get_user_events(platform="IOS", users_list=user_segments["VIP"], timeWindow=defaulTimeWindow):
    filter_obj = create_segment_filter(platform, users_list)
    print("Getting events for {} time interval {}".format(platform, timeWindow))
    result = query.scan(                 \
            datasource=platform, \
            columns=[],   \
            filter=filter_obj,  \
            intervals=[time_window]  
        )
    df = query.export_pandas()
    if not df.empty:
        df["userId"] = df.apply(lambda x: get_valid_userId(x,platform), axis=1)
    return df

def get_sessions_df(event_df,platform):
    # If not in datetime format already:
    event_df['date'] = pd.to_datetime(event_df['timestamp'])
    # sort_values('time') is there just in case it's not already sorted
    event_df['session'] = (event_df.sort_values('date').groupby('userId')['date']
                     .transform(lambda x: (x.diff() > platforms_session_break[platform]).cumsum()+ 1))
    return event_df

def get_sessions_exploration_df(platform,segment,timeWindow=defaulTimeWindow):
    event_df = get_user_events(platform, user_segments[segment])
    session_df = get_sessions_df(event_df, platform=platform)
    session_df = session_df.sort_values(by=['userId','session','date'])
    session_df.reset_index(inplace=True)
    
    session_df['exploration_sequence'] = session_df[['userId','session','properties_log_type','date']].groupby((sample_df['properties_log_type'].shift()!=sample_df['properties_log_type']).cumsum()+1) \
                                 .ngroup()
    exploration_df = session_df[['userId','session','exploration_sequence','properties_log_type','properties_nav_element','properties_event_type','date']]\
                .groupby(['userId','session','exploration_sequence','properties_log_type']).apply(lambda x: pd.Series(
                    {
                        "total_events": x['userId'].count(),
                        "interests": x[x['properties_event_type']=='click']['properties_event_type'].count(),
                        "duration": (x['date'].max() - x['date'].min()),
                        "interactions": x['properties_nav_element'].unique()
                    }
                ))

    
    
    return exploration_df
```
